```python
import numpy as np
import pandas as pd
#paths for downloading the data
import os
#access websites and download content
import urllib
#enable jupyter notebook to render plots in itself
%matplotlib inline
```

```python
url = "https://covid19.who.int/WHO-COVID-19-global-data.csv"
# put the download in a folder called data and a folder called covid
file_path = os.path.join("data", "covid")
```

```python
#directories
os.makedirs(file_path, exist_ok=True)
#path to the data we are downloading
csv_path = os.path.join(file_path, "WHO-COVID-19-global-data.csv")
#going to the url and save it the csv_path
urllib.request.urlretrieve(url, csv_path)
```

```python
# read the csv file
df = pd.read_csv(csv_path)
```

```python
# Death rate in the united states
total_cases = df.loc[df.Country_code == 'US', ['Cumulative_cases']].max().to_numpy()
total_deaths = df.loc[df.Country_code == 'US', ['Cumulative_deaths']].max().to_numpy()
cent = round(float(total_deaths / total_cases) * 100, 2)
print("WHO's data retrieved right now for the US says that, Covid has a {} death rate".format(cent))
pop = 328200000
print("The United States Populaion in 2019 was {}".format(pop))
print("So with a death rate of {}".format(cent))
death_percentage = round(pop * (cent / 100))
print("That means about {} will end up dying".format(death_percentage))
print("We have {} deaths right now".format(int(total_deaths)))
are_we_there = round(float(total_deaths / death_percentage * 100),2)
print("That means we are only {}% of the way there.".format(are_we_there))
```
